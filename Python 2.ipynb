```python
def getDataFromUser():
    D={}
    while True:
        studentID=input("Enter Student ID: ")
        markslist=input("Enter the marks by Comma seperated values: ")
        morestudent=input("Enter no to quit insertion: ")
        if studentID in D:
            print("Student ID is already inserted: ")
        else:
            D[studentID]=markslist.split(",")
        if morestudent.lower()=='no':
            return D
        
            
```

```python
studentData=getDataFromUser()
```

```python
studentData
```

```python
def getAvg(D):
    avgmarks={}
    for x in D:
        L=D[x]
        s=0
        for marks in L:
            s+=int(marks)
            
        avgmarks[x]=s/len(L)
    return avgmarks 
            
            
```

```python
avg=getAvg(studentData)
for x in avg:
    print("Student:",x,"got avg marks as:",avg[x])
```

```python
for idx in range(0,len(countries)):
    C=df3[df3['Country']==countries[idx]].reset_index()
    plt.scatter(np.arange(0,len(C)),C['Cumulative_cases'],color='b',label="Confirmed")
    plt.scatter(np.arange(0,len(C)),C['New_deaths'],color='r',label='Deaths')
    plt.title(countries[idx])
    plt.xlabel('Days since First Suspect')
    plt.ylabel("Number Of Cases")
    plt.legend()
    plt.show()
```
