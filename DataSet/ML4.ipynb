```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

fruits=pd.read_csv('fruit-data-with-colours-master/fruit_data_with_colours.csv')
```

```python
fruits.head()
```

```python
fruit_name=dict(zip(fruits.fruit_name.unique(),fruits.fruit_label.unique()))
fruit_name
```

```python
# plotting a scatter matrix
from matplotlib import cm
from pandas.plotting import scatter_matrix

X = fruits[['height', 'width', 'mass', 'color_score']]
y = fruits['fruit_label']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

cmap = cm.get_cmap('gnuplot')
scatter =scatter_matrix(X_train, c= y_train, marker = 'o', s=40, hist_kwds={'bins':15}, figsize=(9,9), cmap=cmap)
```

```python
from matplotlib import cm
from pandas.plotting import scatter_matrix
X=fruits[['width','height','color_score','mass']]
y=fruits['fruit_label']
X_train , X_test ,y_train , y_test = train_test_split(X , y , random_state = 0)

cmap=cm.get_cmap('gnuplot')
scatter=scatter_matrix(X_train,c=y_train, marker='o',s=40, hist_kwds={'bins':15}, figsize=(9,9),cmap=cmap)
```

```python
from mpl_toolkits.mplot3d import Axes3D
a1=plt.figure()
ax=a1.add_subplot(111,projection='3d')
ax.scatter(X_train['width'],X_train['height'],X_train['mass'],c=y_train,s=100,marker='o')
ax.set_xlabel('width')
ax.set_ylabel('height')
ax.set_zlabel('mass')
plt.show()
```

```python
X=fruits[['width','height','mass']]
y=fruits['fruit_label']
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(X_train,y_train)
```

```python
knn.score(X_test,y_test)
```

```python
fruit_prediction=knn.predict([[4.3,45,23]])
fruit_name[fruit_prediction[0]]
```
