```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
fruits=pd.read_csv('fruit-data-with-colours-master/fruit_data_with_colours.csv')
```

```python
fruits.head()
```

```python
a=dict(zip(fruits.fruit_label.unique(),fruits.fruit_name.unique()))
a
```

```python
from matplotlib import cm
from pandas.plotting import scatter_matrix
X=fruits[['mass','width','height','color_score']]
y=fruits['fruit_label']
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0)

cmap=cm.get_cmap('gnuplot')
c1=scatter_matrix(X_train,c=y_train,marker='o',s=40,hist_kwds={'bins':15},figsize=(9,9),cmap=cmap)
```

```python
from mpl_toolkits.mplot3d import Axes3D
a1=plt.figure()
a2=a1.add_subplot(111,projection='3d')
a2.scatter(X_train['width'],X_train['mass'],X_train['height'],c=y_train,marker='o',s=100)
a2.set_xlabel('width')
a2.set_ylabel('mass')
a2.set_zlabel('height')
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 5)
```

```python
knn.fit(X_train,y_train)
```

```python
knn.score(X_test,y_test)
```
