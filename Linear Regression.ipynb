```python
import pandas as pd
from sklearn import linear_model
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import joblib
import matplotlib.pyplot as plt
drugs_data=pd.read_csv('E:/Coding/Github/datasets/archive/drug.csv')
```

```python
drugs_data
```

```python
X=drugs_data[['age','sex','bp','cholesterol','Na_to_K']].values
y=drugs_data[['drug']].values
le=LabelEncoder()
for i in range(len(X[0])):
    X[:,i]=le.fit_transform(X[:,i])
# print(X)
for i in range(len(y[0])):
    y[:,i]=le.fit_transform(y[:,i])
lreg=linear_model.LinearRegression()
plt.scatter(X.T[0],y)
plt.show()
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
model=lreg.fit(X_train,y_train)
predict=model.predict(X_test)
score=lreg.score(X,y)
print("Predicted Values:",predict)
print("Actual Values:",y_test)
print("Score:",score)
print('Coefficient:',lreg.coef_)
print('Intercept:',lreg.intercept_)
```
